```python
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(['A','B','C','D','E','F'])
values = np.array([4,3,2,5,3,1])

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')


plt.bar(categories, values)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories = np.array(['Freshman','Sophomore','Junior','Senior','Graduate','Other'])
values = np.array([100,150,120,130,80,50])
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.pie(values , labels=categories , autopct='%1.1f%%', explode=[0.2,0,0,0,0,0], colors=colors , shadow=True, startangle=140)
plt.title('Distribution of Students by Year')
plt.show()
```

```python
#scatter graphs
# shows the relationship between two variables , helps to identify a correlation between them (positive , negative , or no correlation)
#Example : relationship between hours studied and exam scores
import matplotlib.pyplot as plt
import numpy as np
hours_studied = np.array([1,2,3,4,5,6])
exam_scores = np.array([50,60,70,80,90,95])
x2 = np.array([1,2,3,4,5,6])
y2 = np.array([55,65,75,85,95,100])

plt.scatter(hours_studied, exam_scores, color='blue', marker='o')
plt.scatter(x2, y2, color='red', marker='s')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Scatter Plot of Hours Studied vs Exam Scores')
plt.show()      
```

```python
#Histogram = A Visual representation of the distribution of a dataset. It groups data into bins and shows the frequency of data points in each bin.
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(loc=80, scale=50,size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=20,
         color='lightgreen', edgecolor='black')

plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
#Subplots = A way to create multiple plots in a single figure, allowing for easy comparison and visualization of different datasets or aspects of the same dataset.
#Figure = The overall container for all the plots and elements in a visualization. It can contain one or more subplots, as well as titles, legends, and other annotations.
#Ax= The individual plot or graph within a figure. Each subplot is represented by an Ax object, which can be customized with titles, labels, and other properties.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3 , 4, 5])



figure, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, x**2, color='blue', marker='o')
axes[0, 1].plot(x, x**3, color='red', marker='s')
axes[0, 0].set_title('x vs x^2 ')
axes[0, 1].set_title('x vs x^3')

plt.show()
```
